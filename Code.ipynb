```python
import pandas as pd

# Load the dataset
df = pd.read_excel(r'Masters_degree_Dataset.xlsx')

def user_preference():
    field_study = {'1': "Computer Science", '2': "Doctoral", '3': "Management", '4': "Other"}
    category = {
                "a": "Tier of the university",
                "b": "Budget Constraint",
                "c": "Exam Requirement",
                "d": "Climate Preferences",
                "e": "Other"
                }
    sub_category_name = {}
    user_preference_dict = {}

    while True:
        user_input1 = input("Hello! What field of study do you want to pursue your masters? (1/2/3/4)\n1. Computer Science\n2. Doctoral\n3. Management\n4. Other\n")

        if user_input1 in ['1', '2', '3', '4']:
            break
        else:
            print("Invalid choice! Please choose among the options provided.")

    if user_input1 in ['1', '2', '3']:
        while True:
            user_input2_1 = input("Please select your preference: (a/b/c/d/e) \na) Tier of the university \nb) Budget Constraint \nc) Exam Requirement \nd) Climate Preferences \ne) Other\n").lower()

            if user_input2_1 in category.keys():
                if user_input2_1 == "a":
                    sub_category_name = {"a": "Tier 1", "b": "Tier 2", "c": "Tier 3"}
                    while True:
                        preference = input("Choose the tier of the university: (a/b/c) \na. Tier 1 \nb. Tier 2 \nc. Tier 3 \n").lower()
                        if preference in sub_category_name.keys():
                            break
                        else:
                            print("Invalid preference! Please select the right option.")
                elif user_input2_1 == "b":
                    sub_category_name = {"a": "<30,000", "b": "<50,000", "c": ">50,000"}
                    while True:
                        preference = input("What is your finance budget? (a/b/c) \na) <30,000 \nb) <50,000 \nc) >50,000 \n").lower()
                        if preference in sub_category_name.keys():
                            break
                        else:
                            print("Invalid preference! Please select the right option.")
                elif user_input2_1 == "c":
                    sub_category_name = {"a": "GMAT", "b": "GRE"}
                    while True:
                        preference = input("Choose your preferred exam: (a/b) \na) GMAT \nb) GRE \n").lower()
                        if preference in sub_category_name.keys():
                            break
                        else:
                            print("Invalid preference! Please select the right option.")
                elif user_input2_1 == "d":
                    sub_category_name = {"a": "< 10° Celsius", "b": "< -10° Celsius", "c": "> -10° Celsius"}
                    while True:
                        preference = input("What is your climate preference? (a/b/c) \na) < 10° Celsius \nb) < -10° Celsius \nc) > -10° Celsius \n").lower()
                        if preference in sub_category_name.keys():
                            break
                        else:
                            print("Invalid preference! Please select the right option.")
                elif user_input2_1 == "e":
                    preference = input("Please specify your concern/preference.\n")
                    print("Thank you! We will get back to you with the details you would like to explore.")
                    return None  
                break

            else:
                print("Invalid preference! Please select the right option.")
                continue

    elif user_input1 == '4':
        user_input2_2 = input("Please specify your field of study.")
        print("Thank you! We will get back to you with the details you would like to explore.")
        preference = user_input2_2

    # Create the user_preference_dict based on the user's choices
    if user_input1 in field_study.keys():
        if user_input1 in ['1', '2', '3']:
            if user_input2_1 in category.keys():
                if preference in sub_category_name.keys():
                    user_preference_dict = {
                        "Field of Study": field_study[user_input1],
                        "Category": category[user_input2_1],
                        "Preference": sub_category_name[preference]}
                else:
                    user_preference_dict = {
                        "Field of Study": field_study[user_input1],
                        "Category": category[user_input2_1],
                        "Preference": preference
                    }
        else:
            user_preference_dict = {
                "Field of Study": field_study[user_input1],
                "Category": user_input2_2
            }

    return user_preference_dict


def filter_universities(preferences, df):
    print(f"Filtering universities for the following preferences: {preferences}")
    # Filter the universities based on user preferences
    filtered_df = df[df['field_of_study'] == preferences['Field of Study']]
    print(f"Number of universities after filtering by field of study: {len(filtered_df)}")

    if 'Category' in preferences:
        if preferences['Category'] == 'Tier of the university':
            filtered_df = filtered_df[filtered_df['Tier'] == preferences['Preference']]
        elif preferences['Category'] == 'Budget Constraint':
            filtered_df = filtered_df[filtered_df['Budget'] == preferences['Preference']]
        elif preferences['Category'] == 'Exam Requirement':
            filtered_df = filtered_df[filtered_df['Exam'] == preferences['Preference']]
        elif preferences['Category'] == 'Climate Preferences':
            filtered_df = filtered_df[filtered_df['Climate'] == preferences['Preference']]

    print(f"Number of universities after applying all filters: {len(filtered_df)}")
    return filtered_df


def main():
    while True:
        user_preference_dict = user_preference()

        # If the user selected "Other" and provided a concern, end the program.
        if user_preference_dict is None:
            break

        print(f"We appreciate your valuable time.")

        filtered_universities = filter_universities(user_preference_dict, df)
        if filtered_universities.empty:
            print("No universities match your preferences.")
        else:
            print("\nHere are the universities that match your preferences:\n")
            print(filtered_universities[['Name', 'Program', 'Tier', 'Budget', 'Exam', 'Climate']])

        # Option to adjust criteria or exit
        user_input = input("\nWould you like to adjust your criteria or exit? (adjust (1) /exit (0)): ").lower()
        if user_input == "0":
            print("Thank you for using the university selection tool! Goodbye!")
            break
        elif user_input != "1":
            print("Invalid choice! Exiting the program.")
            break


if __name__ == '__main__':
    main()
```
